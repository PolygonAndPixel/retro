```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import cPickle as pickle
```

```python
with open('/home/peller/retro_output/scans/downgoing_muon_z-300_evtidx0_newcode/t_-100-100:1_z_-400--200:1/scan.pkl', 'rb') as f:
    scan = pickle.load(f)
```

```python
X, Y = np.meshgrid(scan['scan_values'][0], scan['scan_values'][3])
plt.figure(figsize=(10,8))
test = scan['metrics'][0][:,0,0,:,0,0,0,0].T
cs = plt.contour(X, Y, -test, 15, linewidths = 0.5, colors = 'k')
plt.clabel(cs, inline=1, fontsize=9, fmt='%i')
pc = plt.pcolormesh(X, Y, -test, cmap='Spectral_r')
plt.colorbar(pc)
```

```python
X, Y = np.meshgrid(scan['scan_values'][1], scan['scan_values'][3])
plt.figure(figsize=(10,8))
ref = refscan['metrics'][0][0,:,0,:,0,0,0,0].T
test = scan['metrics'][0][0,:,0,:,0,0,0,0].T
diff = ref - test
diff /= ref
vmax = np.max([np.max(diff),-np.min(diff)])
pc = plt.pcolormesh(X, Y, diff, vmin=-vmax, vmax=vmax, cmap='RdBu')
plt.colorbar(pc)
bestref = np.where(ref==np.min(ref))
plt.scatter(X[bestref], Y[bestref],c='g',marker='o')
best = np.where(test==np.min(test))
plt.scatter(X[best], Y[best],c='r',marker='*')
print X[bestref], Y[bestref]
print X[best], Y[best]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(figsize=(35, 8), ncols=3)

A = uncompressed['metrics'][0][0,0,0,0,:,:,0,0]
B = compressed['metrics'][0][0,0,0,0,:,:,0,0]

X, Y = np.meshgrid(uncompressed['scan_values'][5], uncompressed['scan_values'][4])


pc = ax1.pcolormesh(X, Y, -A, cmap='Spectral_r', vmin=-700)
plt.colorbar(pc, ax=ax1)
ax1.set_title('Original tables')
pc = ax2.pcolormesh(X, Y, -B, cmap='Spectral_r', vmin=-700)
plt.colorbar(pc, ax=ax2)
ax2.set_title('Compressed tables')
#vmax = np.max([np.max(A-B),-np.min(A-B)])
#pc = ax3.pcolormesh(X, Y, A-B, cmap='RdBu', vmin=-vmax, vmax=vmax)
#plt.colorbar(pc, ax=ax3)
vmax = np.max([np.max((A-B)/(A+B)),-np.min((A-B)/(A+B))])
pc = ax3.pcolormesh(X, Y, (A-B)/(A+B), cmap='RdBu', vmin=-vmax, vmax=vmax)
plt.colorbar(pc, ax=ax3)
ax3.set_title('relative difference')
```
